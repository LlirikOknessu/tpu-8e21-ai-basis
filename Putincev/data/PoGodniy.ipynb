```python
import pandas as pd

# Создаем DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'Score': [80, 90, 75]
}
df = pd.DataFrame(data)
print("Исходный DataFrame:")
print(df)

# Используем метод assign() для добавления нового столбца
new_column = df.assign(Result=lambda x: x['Score'] * 2)

# Печатаем результат
print("Новый DataFrame:")
print(new_column)
```

```python
import pandas as pd

# Создаем DataFrame
data = {
    'Category': ['A', 'B', 'C', 'A', 'B', 'C'],
    'Value': [1, 2, 3, 4, 5, 6]
}
df = pd.DataFrame(data)

# Группируем по категории и вычисляем среднее значение
averaged_values = df.groupby('Category').mean()

# Печатаем результат
print("Среднее значение для каждой категории:")
print(averaged_values)
```

```python
import pandas as pd

# Создаём DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'Score': [80, 90, 75]
}
df = pd.DataFrame(data)

# Добавляем новый столбец на основе существующего
new_column = df.assign(Result=lambda x: x['Score'] * 2)

# Печатаем результат
print("Новый DataFrame:")
print(new_column)
```

```python
import pandas as pd

# Создаём DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Address': ['123 Main St', '456 Elm Main', '789 Oak Ave']
}
df = pd.DataFrame(data)

# Выбираем строки, содержащие 'Main' в столбце 'Address'
addresses_with_main = df[df['Address'].str.contains('Main')]

# Печатаем результат
print("Адреса, содержащие 'Main':")
print(addresses_with_main)
```
